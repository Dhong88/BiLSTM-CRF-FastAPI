```python
from gensim.models import Word2Vec
import numpy as np
import pickle
```

```python
import pandas as pd

file=open(copied_path,"rt")
data=[]
contents =file.readlines()
indx=1
for x in contents:
    words=x.split()
    #print(len(words))
    for i in range(len(words)):
      tags=words[i].split("/")
      #print(len(tags))
      data.append([indx,tags[0],tags[1]])
    indx+=1



df=pd.DataFrame(data,columns=['sentence_idx','word','tag'])
vocab=df['word'].unique()
```

```python
word=np.array(vocab)
print(word)
```

```python
wordtuple=list(word)
w=[wordtuple]
print(w)
```

```python
modelgensim=Word2Vec(sentences=w,min_count=1,size=300)
```

```python
modelgensim.save('modelPGKMalayTweet.bin')
```
